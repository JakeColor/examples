## B: Which Team Metrics Predict Victory?

We'll compare and contrast 4 subsets of the data: Home team winners + losers and Away team winners + losers. The main goal of this section is identifying factors (across both Home and Away metrics) important to predicting victory for a Home team. The selection of metrics is done with a variety of statistical machine learning methods.

Let's isolate the design matrix and response vector, a boolean column indiciating Home winners. `Sklearn` requires `NumPy` arrays for model fitting.

```python
train_design_cols = train.columns.values[7:]
design = train[train_design_cols].values
design.shape
```

```python
H_team_win = (train["H-Score"] > train["A-Score"]).astype(bool)
print(np.bincount(H_team_win))
```

```python
sum(H_team_win) / len(H_team_win)
```

Over the last 12 seasons, Home teams have won 57% of regular season games. Any investigation or model of our NFL data should predict home winners at an accuracy significantly better than this baseline rate.

### i)    Student's <i>t</i>-Test

As a jump-off point, we'll test metrics for significant differences in mean between the first pair of subsets: Home team winners + losers. A significant difference may indicate the metric is important. For example, the average of every Home winner's <code>H-ScoreOff</code> metric will be compared to the same average <code>H-ScoreOff</code> for Home losers. These comparisons will be made for every statistic in the design matrix.

Although many performance metrics in the training matrix are not normally distributed, I chose to apply the popular Student's <i>t</i>-test to compare the sample means of winners and losers metrics because our large sample sizes mean the $t$ statistic will be a very good approximation.

Also, The two sample $t$-test assumes the observations are independent, which in our case is not completely true because NFL teams in the same division play one another twice. However, these rematches are a small portion of the total training set and the non-independent data (the first game between the teams) is in turn a small portion of each team's metrics.

```python
from scipy.stats import ttest_ind

ALPHA = .001
t_sig_mets = []
for met in train_design_cols:
    p_value = ttest_ind(train[H_team_win][met], train[~H_team_win][met])[1]
    
    if p_value < ALPHA:
        t_sig_mets.append(met)
        
t_sig_mets[1:5]
```

```python
len(t_sig_mets)
```

Even at an $\alpha$ of 1%, 32 predictor means are significantly different for the Home winners + losers subpopulations. However, if we plot the respective distributions for the first significant metric, the overlap is more significant than we would like to see.

For a more thorough tree-based approach to feature selection, we can fit a Random Forest of trees to the dataset using the `RandomForestClassifier` module in `sklearn`. The method fits hundreds of decorrelated trees to bootstrapped subsamples of the dataset and predicts classifications by average estimated probability across the trees. Relative feature importance is gauged by <i>estimating</i> the percentage of samples each metric contributes to classification.

__Note:__ This code is based on the example provided by `sklearn`'s `ExtraTreeClassifier` online documentation, which can be found [here](http://scikit-learn.org/stable/modules/generated/sklearn.ensemble.ExtraTreesClassifier.html).

### 1) Turnovers Matter (Pt. I)

The branches on our decision tree that lead to this observation are:

1. $H\_RushYdsDef > 115.39 \frac{yards}{game}$
1. $H\_ThirdDownPctOff > 34.795\%$
1. $A\_FirstDownOff \leq 16.085 \frac{1st\,downs}{game}$
1. $H\_PassIntDef > .795 \frac{interceptions}{game}$

__Home Team Record__: 50-7

The dominance of the Seattle Seahawks' vaunted "Legion of Boom" defense may have a root in our historical data. The speedy and physical Seattle legionairres, known for their athletic interceptions, create turnovers by the dozen- the team is ranked in the top 5 for turnover margin every season since the current Legion members joined the Seahawks in 2012.  The team played in the Super Bowl twice over that timespan.

How well would a team do emulating the Seahawks' ability to create turnovers?  According to the listed branches, even Home teams with a weak run defense (branch #1) won 50 out of 57 games provided they average more than .795 interceptions per game (#2) AND their opponents average relatively few plays per game (#2, #3).

__Insight__

Wary of correlation not implying causation,* actionable advice for NFL teams may be scouting of taller, more athletic defensive backs than the current norm. I am <i>not</i>, however, suggesting sending hordes of DBs onto the field to gamble on interceptions every play.  Such a strategy would most likely lift the team's $H\_RushYdsDef$ far, far above 115 ypg... and a lot of fired coaches.

*It is common mispractice to confuse a relationship between variables as an actual causal effect.  In statistical parlance, this mistake is known as "correlation does not imply causation".

__Note:__ This code is based on the example provided by `sklearn.feature_selection.RFECV`'s online documentation, which can be found [here](http://scikit-learn.org/stable/modules/generated/sklearn.feature_selection.RFECV.html)

```python
def lsvc_feature_selection(design, response, feature_names, verbose=0):
    """ Uses the Linear Support Vector Classifier algorithm to select
        the optimal number of features for the given classification
        problem.
        
        The problem is defined by an input DESIGN matrix with associated
        FEATURE_NAMES and a RESPONSE vector. The auxiliary VERBOSE flag
        allows the user to control output of optimal number of features
        and a learning curve plot.
        
        Returns an array of selected features, RFECV_SIG_METS.
    """
    lsvc = LinearSVC()
    rfecv = RFECV(estimator=lsvc, step=1, cv=StratifiedKFold(response,
                                                             n_folds = 5),
                  scoring='accuracy')
    rfecv.fit(design, response)
    
    if verbose == 1:
        print("Optimal number of features : %d" % rfecv.n_features_)
        plt.figure()
        plt.xlabel("Selected Features")
        plt.ylabel("Accuracy Score")
        plt.plot(range(1, len(rfecv.grid_scores_) + 1), rfecv.grid_scores_)
        plt.show()
        
    rfecv_sig_mets = (rfecv.ranking_ == 1)
    rfecv_sig_mets = feature_names[rfecv_sig_mets]
    
    return rfecv_sig_mets
```

```python
def lasso_feature_selection(design, response, feature_names, sd=1, verbose=0):
    """ Uses the L1 regularization norm with logistic regression for feature selection.
    
        Classification problem is defined by an input (scaled) DESIGN matrix with
        associated FEATURE_NAMES and a RESPONSE vector.
        
        Auxiliary VERBOSE flag allows user to suppress or allow output learning curve.
        
        Selects features by fitting model with minimum feature count with accuracy
        within SD standard deviations of maximum score.
        
        Returns an array of selected features, FEATURES.
    """
    #define search grid...
    search = np.linspace(-3, 1, num=25)
    non_zeros = np.empty(25)
    scores = np.empty(25)
    
    for i, C in enumerate(10 ** search):
        lasso_log = LogisticRegression(C=C, penalty = 'l1')
        lasso_log.fit(reg_design, response)
        sparsity = sum(lasso_log.coef_.ravel() == 0)
        sparsity = (int(reg_design.shape[1]) - sparsity)
        score = lasso_log.score(reg_design, response)
        
        non_zeros[i] = sparsity
        scores[i] = score
        
    if verbose == 1:
        print("Number of significant features : %d" % sparsity)
        plt.figure()
        plt.title("")
        plt.xlabel("Log of Regularization Parameter (C)")
        plt.ylabel("Accuracy")
        plt.scatter(search, scores)
        plt.plot(search, scores)
        
        for label, x, y in zip(non_zeros, search, scores):
            plt.annotate(str(int(label)), xy = (x,y),
                         ha = 'right', va = 'bottom')
        plt.show()
    
    # isolate optimal number of features w.r.t. SD rule...
    max_score = max(scores)
    non_triv_std = np.std(scores[np.where(non_zeros > 0)[0]])
    min_loc = min(np.where(scores > max_score - sd * non_triv_std)[0])
    
    # refit and return selected features...
    C = 10 ** search[min_loc]
    lasso_log = LogisticRegression(C=C, penalty = 'l1')
    lasso_log.fit(reg_design, response)
    
    l_sig_mets = (lasso_log.coef_.ravel() != 0)
    l_sig_mets = feature_names[l_sig_mets]
    
    return l_sig_mets
```

```python
def top_corrs(metric, n):
    """ Returns a list of the top N correlations of METRIC with columns in DF.
    """
    corrs = []
    for i, met in enumerate(train_design_cols):
        corrs.append((met, np.corrcoef(train[metric], train[met])[1,0]))
        
    corrs = sorted(corrs, key = lambda x: x[1], reverse = True)
    #first returned will be metric itself with 1.0 corr...
    return corrs[1:n+1]
```

```python
def corrs_analyze(met, met_corrs, col = "dodgerblue"):
    mets = [tup[0] for tup in met_corrs]
    cor = [tup[1] for tup in met_corrs]
    n = len(mets)
    fig = plt.figure(figsize=(18, 6))
    fig.suptitle(met + " vs. ...", fontsize = 20)
    plt.subplot(1, n, 1)
    
    for i in range(n):
        sub = fig.add_subplot(1, n, i)
        sub.scatter(train[met], train[mets[i]], alpha = .5, color = col)
        sub.set_title(mets[i] + ": " + str(round(cor[i], 3)))
    
    plt.show()
```
